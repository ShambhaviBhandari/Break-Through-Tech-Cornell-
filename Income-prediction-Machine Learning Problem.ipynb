```python
# YOUR CODE HERE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score
import matplotlib.pyplot as plt
```

```python
# YOUR CODE HERE

# One-hot encoded education features, orginally also did sex but took it off because it was not neccsiary 

features = ['education_Bachelors', 'education_Doctorate','education_Masters', 'education_Prof-school','education_Some-college']

# I did not want income_binary in text form I wanted to convert it to binary numbers 0 for <=50k and 1 for >=50k
# this is because I am doing binary classification for my logistic regression model and this was causing me alot of errors before I converted.
if data['income_binary'].dtype == 'object':
    data['income_binary'] = data['income_binary'].map({'<=50K': 0, '>50K': 1})

# to check that data is being grouped properly for both classes 
print(data[['sex_selfID', 'income_binary']].value_counts(dropna=False))

# This is where I came in with my second attempt to train my model because I wanted two groups one for Female and one for Non-Female
#separate models for each group both using logistic regression
gender_groups = {'Female': data[data['sex_selfID'] == 'Female'],'Non-Female': data[data['sex_selfID'] == 'Non-Female']}

#this loops through each gender group 
for label, group_data in gender_groups.items():
    group_data = group_data.dropna(subset=['income_binary'])  # drop rows with missing values because this can cause an error in the model
    
#I had trial and error with this: this is so if there is no data left after removing values it will skip the group
    if group_data.shape[0] == 0:
        print(f"\n{label.upper()} MODEL: skip")
        continue
#input and target value
    X = group_data[features]
    y = group_data['income_binary']
# in case everyon has the same income we can't train the model
    if len(y.unique()) < 2:
        print(f"\n{label.upper()} MODEL: skip")
        continue

    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=80)

    # train model: logistic regression 
    model = LogisticRegression(max_iter=300)
    model.fit(X_train, y_train)

    # Predict
    y_pred = model.predict(X_test)

    # Evaluate and print for the model
    print(f"\n{label.upper()} MODEL:")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred, average='binary'))
    print("Recall:", recall_score(y_test, y_pred, average='binary'))
```
